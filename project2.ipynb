```python
import seaborn as sb
sb.get_dataset_names()
```

```python
import matplotlib.pyplot as plt
t=sb.load_dataset('taxis')
t.head(20)
```

```python
plt.figure(figsize=(6,3))
sb.countplot(x='dropoff_borough',data=t)
plt.title('Dropoff cities')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sb.barplot(x='payment',y='fare',data=t,color='green')
plt.title('Payment method by distance')
plt.xlabel('payment')
plt.ylabel('distance')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sb.histplot(t['fare'],bins=20,kde=True,color='blue')
plt.title('Fare amount distribution')
plt.show()
```

```python
counts=t['pickup_borough'].value_counts()
counts
```

```python
pickup_counts=t['pickup_borough'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(pickup_counts,labels=['Manhattan','Queens','Brooklyn','Bronx'],colors=['red','green','blue','pink'],autopct='%1.1f%%')
plt.title('Pickup cities')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sb.boxplot(x='passengers',data=t)
plt.title('Distribution of passenger count')
plt.xlabel('')
plt.ylabel('passengers')
plt.show()
```

```python
sb.scatterplot(x='distance',y='fare',data=t)
plt.title('Fare vs Distance for travel')
plt.xlabel('Distance')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sb.scatterplot(x='distance',y='fare', hue='dropoff_borough',data=t)
plt.title('Distance vs Fare with pickup city')
plt.xlabel('Distance')
plt.ylabel('Fare')
plt.legend(title='pickup city')
plt.show()
```

```python
numerical=t[['passengers','distance','fare','tip','tolls','total']].corr()
plt.figure(figsize=(4,4))
sb.heatmap(numerical,annot=True,cmap='rainbow')
plt.title('Correlation heatmap of taxis')
plt.show()
```

```python
plt.figure(figsize=(7,4))
sb.violinplot(x='color',y='passengers',hue='payment',data=t,split=True)
plt.title('Passenger distribution by taxi color and payment mode')
plt.xlabel('color')
plt.ylabel('passengers')
plt.legend(title='payment')
plt.show()
```

```python
sb.kdeplot(data=t['distance'])
plt.title('Taxi distance distribution')
plt.xlabel('Distance')
plt.show()
```
