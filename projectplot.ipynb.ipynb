```python
import pandas as pd
t=pd.read_csv("/content/loan.csv")
t
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Count Plot


```python
plt.figure(figsize=(10,4))
sns.countplot(x='Gender',data=t,color='green')
plt.title('Gender by count')
plt.show()
```

Bar Plot


```python
plt.figure(figsize=(10,4))
sns.barplot(x='Loan_Status',y='ApplicantIncome',data=t,color='red')
plt.title('Loan status by Applicant income')
plt.xlabel('Loan_Status')
plt.ylabel('ApplicantIncome')
plt.show()
```

Histogram

```python
plt.figure(figsize=(8,4))
sns.histplot(t['LoanAmount'],bins=20,kde=True,color='blue')
plt.title('LoanAmount distribution')
plt.show()
```

Pie Plot


```python
education_counts=t['Education'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(education_counts,labels=['Graduate','Not Graduate'],colors=['blue','green'],autopct='%1.1f%%')
plt.title('Education')
plt.show()
```

Box Plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='LoanAmount',data=t)
plt.title('LoanAmount distribution')
plt.xlabel('')
plt.ylabel('LoanAmount')
plt.show()
```

Scatter Plot

```python
sns.scatterplot(x='ApplicantIncome',y='LoanAmount',data=t)
plt.title('Applicant income vs Loan amount')
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='ApplicantIncome',y='LoanAmount', hue='Education',style='Gender',data=t)
plt.title('Applicant Income vs Loan Amount with education and gender')
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.legend(title='Survived')
plt.show()
```

Heat Map

```python
numerical=t[['ApplicantIncome','CoapplicantIncome','LoanAmount']].corr()
plt.figure(figsize=(4,4))
sns.heatmap(numerical,annot=True,cmap='rainbow')
plt.title('Correlation heatmap of Loan details of Customer')
plt.show()
```

Violin Plot

```python
plt.figure(figsize=(7,4))
sns.violinplot(x='Property_Area',y='LoanAmount',hue='Education',data=t,split=True)
plt.title('Loan amount distribution by property area and education')
plt.xlabel('Property Area')
plt.ylabel('Loan Amount')
plt.legend(title='Education')
plt.show()
```

KDE Plot

```python
sns.kdeplot(data=t['LoanAmount'])
plt.title('Loan Amount distribution')
plt.xlabel('Loan Amount')
plt.show()
```

Pair Plot

```python
sns.pairplot(t)
```

```python
sns.pairplot(t[['ApplicantIncome','LoanAmount','Property_Area','CoapplicantIncome']])
```

Swarm Plot

```python
plt.figure(figsize=(7,4))
sns.swarmplot(x='Property_Area',y='LoanAmount',hue='Education',palette='rainbow',data=t,dodge=True)
plt.title('Loan amount distribution by property area and education')
plt.xlabel('Property_Area')
plt.ylabel('LoanAmount')
plt.legend(title='Education')
plt.show()
```
